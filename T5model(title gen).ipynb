```python
import pandas as pd
df = pd.read_csv('articles1.csv')
```

```python
df = df[['title','content']]
```

```python
df = df[:3500]
```

```python
papers = df
papers = papers[['title','content']]
papers.columns = ['target_text','input_text']
```

```python
papers = papers[:100]
```

```python
len(papers)
```

```python
eval_df = papers.sample(frac=0.2, random_state = 42)
train_df = papers.drop(eval_df.index)
```

```python
len(eval_df
  )
```

```python
!pip install simpletransformers
```

```python
import logging
from simpletransformers.t5 import T5Model
import pandas as pd

logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)

train_df['prefix'] = "summarize"
eval_df['prefix'] = "summarize"


model_args = {
    "reprocess_input_data": True,
    "overwrite_output_dir": True,
    "max_seq_length": 512,
    "train_batch_size": 16,
    "num_train_epochs": 4,
}

# Create T5 Model
model = T5Model('t5',"t5-small", args=model_args,use_cuda = True)

# Train T5 Model on new task
model.train_model(train_df)

# Evaluate T5 Model on new task
results = model.eval_model(eval_df)

# Predict with trained T5 model
#print(model.predict(["convert: four"]))
```
